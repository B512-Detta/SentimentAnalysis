```python
import pandas as pd
import re
import string
import nltk
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from tqdm import tqdm


import numpy as np # untuk mengolah array
from sklearn.model_selection import train_test_split # untuk membagi dataset / data training dan test
from sklearn.metrics import accuracy_score # untuk mengetahui nilai akurasi
from sklearn.metrics import precision_score # untuk mengetahui nilai precision
from sklearn.metrics import recall_score # untuk mengetahui nilai recall
from sklearn.metrics import confusion_matrix # untuk melihat evaluasi model dengan confusion matrix
from sklearn.naive_bayes import MultinomialNB

import matplotlib.pyplot as plt # untuk memperlihatkan visualisasi data
import seaborn as sns # untuk memperlihatkan visualisasi data
import missingno # digunakan untuk mengetahui apakah terdapat nilai null atau tidak
import matplotlib.pyplot as plt


from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Membuat fungsi membersihkan tweet
def clean_tweet(Text):
    # Menghapus RT dan karakter setelahnya sampai titik dua
    Text = re.sub(r'RT .*?:', '', Text)
    # Menghapus URL
    Text = re.sub(r'http\S+|www.\S+', '', Text)
    # Menghapus username (teks yang dimulai dengan @)
    Text = re.sub(r'@[A-Za-z0-9_]+', '', Text)
    # Menghapus karakter yang bukan huruf dan spasi
    Text = re.sub(r'[^a-zA-Z ]', '', Text)
    # Menghapus spasi berlebih
    Text = re.sub(r'\s+', ' ', Text).strip()
    return Text

def casefoldingText(text): # mengubah semua karakter dalam teks menjadi huruf kecil
    text = text.lower()
    return text

def tokenizingText(text): # tokenisasi atau pemisahan kata
    text = word_tokenize(text)
    return text

def filteringText(text): # menghilangkan stopwors dalam teks
    listStopwords = set(stopwords.words('indonesian'))
    filtered = []
    for txt in text:
        if txt not in listStopwords:
            filtered.append(txt)
    text = filtered
    return text


def toSentence(list_words):
    sentence = ' '.join(word for word in list_words)
    return sentence

# Gunakan fungsi apply untuk membersihkan setiap tweet
df['Text'] = df['Text'].fillna('')
df['text_clean'] = df['Text'].apply(clean_tweet)
df['text_clean'] = df['text_clean'].apply(casefoldingText)
df['text_clean'] = df['text_clean'].apply(tokenizingText)
df['text_clean'] = df['text_clean'].apply(filteringText)
#df['text_clean'] = df['text_clean'].apply(stemmingText)
df['text_clean'] = df['text_clean'].apply(toSentence)

print(df.head(10))
```

```python
df = df.dropna(subset=['text_clean', 'Sentiment'])

X_train = df['text_clean']
y_train = df['Sentiment']

print(X_train.shape, y_train.shape)
```

```python
#splitting data
# proses membagi data menjadi 2 yaitu data train dan data test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['text_clean'],
                                                    df['Sentiment'],
                                                    random_state=21,
                                                    test_size = 0.3
                                                    )

# memperlihatkan hasil dari splitting data

print("X_train_shape : ", X_train.shape)
print("X_test_shape : ", X_test.shape)
print("X_train_shape : ", y_train.shape)
print("X_test_shape : ", y_test.shape)
```

```python
# pemrosesan pembobotan tf-idf
# pembobotan pada suatu kata
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

tf = TfidfVectorizer(max_df=0.8, min_df=3, ngram_range=(1, 3))
tf.fit(df['text_clean'])

X_tra = tf.transform(X_train)
X_te = tf.transform(X_test)
```

```python
# penggunaan model naive bayes
# memprediksi dengan model naive bayes

from sklearn.naive_bayes import MultinomialNB

mnb = MultinomialNB(alpha=0.3)
mnb.fit(X_tra, y_train)
```

```python
# Memprediksi dengan model terbaik
pred1 = mnb.predict(X_te)
```

```python
#evaluasi, dan menampilkan hasil
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

print(classification_report(y_test, pred1))

print(confusion_matrix(y_test, pred1))
```

```python
# untuk memperlihatkan visualisasi data
# dengan menggunakan confusion matrix

from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt # untuk memperlihatkan visualisasi data
import seaborn as sns # untuk memperlihatkan visualisasi data
import missingno # digunakan untuk mengetahui apakah terdapat nilai null atau tidak
import matplotlib.pyplot as plt

sns.heatmap(confusion_matrix(y_test, pred1),
            annot=True,
            cmap='viridis',
            fmt='.0f')
plt.xlabel('Predicted Values',
           fontdict={'size':14},
           labelpad=10)
plt.ylabel('Actual Values',
           fontdict={'size':14},
           labelpad=10)
plt.title('Confusion Matrix for Naive Bayes')
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def show_wordcloud(df, title = None):

    text = " ".join([x for x in df])

    wordcloud = WordCloud(
        background_color='black',
        width = 800,
        height = 500,
        max_words=100,
        max_font_size=300,
        scale=3,
        random_state=1
    ).generate(text)

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title:
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud, interpolation='bilinear')
    plt.imshow(wordcloud)
    plt.show()


show_wordcloud(text_clean['text_clean'])
show_wordcloud(text_clean.loc[text_clean['Sentiment'] == 1]['text_clean'])
show_wordcloud(text_clean.loc[text_clean['Sentiment'] == -1]['text_clean'])
show_wordcloud(text_clean.loc[text_clean['Sentiment'] == 0]['text_clean'])
```
